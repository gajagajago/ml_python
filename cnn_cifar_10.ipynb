```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.layers import Conv2D, MaxPool2D
from tensorflow.keras.optimizers import Adam
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
# 데이터 전처리: tensor로 변환, 데이터 normalization (sklearn scaler 사용)
x_train = x_train.reshape(-1, 32, 32, 3) / 255.0
x_test = x_test.reshape(-1, 32, 32, 3) / 255.0
```

```python
# CNN model 구축: 2 conv - pooling - dropout - flatten - hidden - dropout - output
```

```python
# Feature Extractor
cnn_cifar_10 = Sequential()
cnn_cifar_10.add(Conv2D(input_shape=(32,32,3), kernel_size=(3,3), filters=32, activation='relu', padding='same'))
cnn_cifar_10.add(Conv2D(kernel_size=(3,3), filters=32, activation='relu', padding='same'))
cnn_cifar_10.add(MaxPool2D(pool_size=(2,2)))
cnn_cifar_10.add(Dropout(0.25))
```

```python
# Classifier
cnn_cifar_10.add(Flatten())
cnn_cifar_10.add(Dense(128, activation='relu'))
cnn_cifar_10.add(Dropout(0.5))
cnn_cifar_10.add(Dense(10, activation='softmax'))
```

```python
cnn_cifar_10.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn_cifar_10.summary()
```

```python
# 학습
hist = cnn_cifar_10.fit(x_train, y_train, batch_size=128, epochs=30, validation_split=0.2)
```

```python
# evaluate
cnn_cifar_10.evaluate(x_test, y_test)
```

```python
cifar_10_item = {
    0: "airplane",
    1: "automobile",
    2: "bird",
    3: "cat",
    4: "deer",
    5: "dog",
    6: "frog",
    7: "horse",
    8: "ship",
    9: "truck",
}
```
