```python
from konlpy.tag import Okt
okt = Okt()
```

```python
kongjwi_patjwi = open('kongjwi_patjwi.txt', 'r')
```

```python
lines = kongjwi_patjwi.readlines()
word_pos_list = []

exclude_pos_list = ['Punctuation', 'Josa', 'Foreign']
f_skip_exclude_pos = lambda word_pos : word_pos[1] not in exclude_pos_list

for line in lines:
    if (line != '\n'):
        line_word_pos = okt.pos(line, norm=True, stem=True) # norm: 표준화 / stem: 어간 추출
        word_pos_list += list(filter(f_skip_exclude_pos, line_word_pos))
```

```python
print(word_pos_list[:20])
```

```python
get_word = lambda x : x[0]

word_list = list(map(get_word, word_pos_list))
```

```python
print(word_list)
```

```python
kongjwi_patjwi.close()

with open('x_train.txt', 'w', encoding='utf-8') as x_train:
    x_train.write('\n'.join(word_list))
```

```python
from gensim.models import word2vec
```

```python
x_train = open('x_train.txt', 'r')
learn_data = word2vec.LineSentence(x_train)
```

```python
model = word2vec.Word2Vec(learn_data, window=5, hs=1, min_count=2, sg=1)
```
