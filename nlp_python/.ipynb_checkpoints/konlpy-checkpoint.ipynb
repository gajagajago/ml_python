### konlpy

```python
from konlpy.tag import Komoran
komoran = Komoran(userdic='komoran_dict.txt') # Import user dictionary 
```

```python
komoran.morphs('우왕 코모란도 오픈 소스가 되었어요') # morphemes: 형태소
```

```python
komoran.pos('우왕 코모란도 오픈 소스가 되었어요') # part of speech: 품사
```

```python
komoran.morphs('바람과 함께 사라지다를 보고 눈물을 흘렸어요')
```

### beautifulsoup

```python
from bs4 import BeautifulSoup
```

```python
html_doc = """<html><head><title name="main_title">The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""
```

```python
soup = BeautifulSoup(html_doc, 'html.parser') # Parse HTML to Python obj: Tag, NavigableString, BeautifulSoup, Comment
```

```python
print(soup.prettify())
```

```python
print(soup.title.name)
print(soup.title['name'])
soup.title['id'] = '1'
print(soup.title.attrs)
```

### Crawling

```python
import urllib.request
```

```python
url = 'https://cse.snu.ac.kr/about'
url_open = urllib.request.urlopen(url).read()
soup = BeautifulSoup(url_open, 'html.parser')
```

```python
div_main = soup.find('div', id='main')
```

```python
intro = div_main.find_all('div', 'field-items')[1] # get e by class name
```

```python
p_list = intro.find_all('p')
```

```python
for c in p_list:
    print(c.string)
```

### 형태소 분석

```python
nouns = mecab.nouns(intro.get_text())
```

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
count = Counter(nouns)
```

```python
noun_dict = dict(count.most_common())
print(noun_dict)
```

```python
wc = WordCloud(background_color='white', font_path='baskin_robbins.otf')
cloud = wc.generate_from_frequencies(noun_dict)
```

```python
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
